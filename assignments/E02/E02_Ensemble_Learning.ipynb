# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf` ou `ab`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
import time
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

def load_data(seed=42):
    X, y = fetch_openml('Fashion-MNIST', version=1, as_frame=False, return_X_y=True, parser='auto')
    y = y.astype(int)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=seed, stratify=y
    )
    return X_train, X_test, y_train, y_test

def train_random_forest(X_train, y_train, seed=42):
    rf_model = RandomForestClassifier(n_estimators=100, random_state=seed, n_jobs=-1)
    rf_model.fit(X_train, y_train)
    return rf_model

def train_adaboost(X_train, y_train, seed=42):
    ada_model = AdaBoostClassifier(n_estimators=50, random_state=seed)
    ada_model.fit(X_train, y_train)
    return ada_model

def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    return accuracy_score(y_test, y_pred)

def run_pipeline(model_type="rf", seed=42):
    X_train, X_test, y_train, y_test = load_data(seed=seed)
    
    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed=seed)
    elif model_type == "ab":
        model = train_adaboost(X_train, y_train, seed=seed)
    else:
        raise ValueError("model_type deve ser 'rf' ou 'ab'")
        
    acc = evaluate(model, X_test, y_test)
    return acc

if __name__ == "__main__":
    acc_rf = run_pipeline(model_type="rf", seed=42)
    print(f"Resultado Final RF: {acc_rf:.4f}")
    
    acc_ab = run_pipeline(model_type="ab", seed=42)
    print(f"Resultado Final AB: {acc_ab:.4f}")
````

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

O overfitting começa quando a acurácia de treino sobe e a de teste cai , por volta da profundidade 4.

Com max_depth=None, a árvore atinge 100% no treino, porque o algoritmo continua criando divisões até que cada folha seja pura (tenha apenas uma classe).

# Questão 5

Execute o pipeline para ambos os modelos:

- Random Forest
- AdaBoost

## Apresente:
- Acurácia, Precisão, Recall e F1-Score de cada modelo

## Responda:
- Qual modelo apresentou melhor desempenho inicial?

**Solução**:

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

def evaluate_complete(model, X_test, y_test):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    return {
        "Acurácia": acc,
        "Precisão": prec,
        "Recall": rec,
        "F1-Score": f1
    }

if __name__ == "__main__":
    SEED = 42
    X_train, X_test, y_train, y_test = load_data(seed=SEED)
    modelo_rf = train_random_forest(X_train, y_train, seed=SEED)
    metrics_rf = evaluate_complete(modelo_rf, X_test, y_test)
    modelo_ada = train_adaboost(X_train, y_train, seed=SEED)
    metrics_ada = evaluate_complete(modelo_ada, X_test, y_test)
    df_metrics = pd.DataFrame([metrics_rf, metrics_ada], index=["Random Forest", "AdaBoost"])
    print("\n--- COMPARATIVO DE DESEMPENHO ---")
    print(df_metrics.to_markdown())
````

# Questão 6

Execute o pipeline utilizando diferentes seeds (ex: 42 e 7).

## Analise:
- Os resultados mudaram?

## Responda:
- O experimento é reprodutível? Justifique.

1. Os resultados mudaram?
Sim, os resultados mudam levemente entre a seed=42 e a seed=7. Isso ocorre porque a semente controla como os dados são embaralhados e como a Random Forest seleciona aleatoriamente as características de cada árvore.

2. O experimento é reprodutível? Justifique.
Sim. Pois ao utilizar o mesmo código e a mesma semente, o resultado será idêntico.

**Solução**:

````python
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

def load_data(seed=42):
    X, y = fetch_openml('Fashion-MNIST', version=1, as_frame=False, return_X_y=True, parser='auto')
    y = y.astype(int)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=seed, stratify=y
    )
    return X_train, X_test, y_train, y_test

def train_and_evaluate(X_train, X_test, y_train, y_test, model_type="rf", seed=42):
    if model_type == "rf":
        model = RandomForestClassifier(n_estimators=100, random_state=seed, n_jobs=-1)
    else:
        model = AdaBoostClassifier(n_estimators=50, random_state=seed)
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return accuracy_score(y_test, y_pred)

def run_experiment(seeds=[42, 7]):
    results = []
    
    for s in seeds:
        print(f"Processando Seed: {s}...")
        X_train, X_test, y_train, y_test = load_data(seed=s)
        acc_rf = train_and_evaluate(X_train, X_test, y_train, y_test, "rf", s)
        results.append({"Seed": s, "Modelo": "Random Forest", "Acurácia": acc_rf})
        acc_ab = train_and_evaluate(X_train, X_test, y_train, y_test, "ab", s)
        results.append({"Seed": s, "Modelo": "AdaBoost", "Acurácia": acc_ab})
    
    df = pd.DataFrame(results)
    print("\n" + "="*30)
    print("RESULTADOS DO EXPERIMENTO")
    print("="*30)
    print(df.to_string(index=False))
    return df

if __name__ == "__main__":
    run_experiment()
````

````python
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

def load_data(seed=42):
    X, y = fetch_openml('Fashion-MNIST', version=1, as_frame=False, return_X_y=True, parser='auto')
    y = y.astype(int)
    return train_test_split(X, y, test_size=0.2, random_state=seed, stratify=y)

if __name__ == "__main__":
    SEED = 42
    X_train, X_test, y_train, y_test = load_data(seed=SEED)

    rf = RandomForestClassifier(n_estimators=100, random_state=SEED, n_jobs=-1)
    rf.fit(X_train, y_train)
    
    rf_train_acc = accuracy_score(y_train, rf.predict(X_train))
    rf_test_acc = accuracy_score(y_test, rf.predict(X_test))
    ada = AdaBoostClassifier(n_estimators=50, random_state=SEED)
    ada.fit(X_train, y_train)
    
    ada_train_acc = accuracy_score(y_train, ada.predict(X_train))
    ada_test_acc = accuracy_score(y_test, ada.predict(X_test))
    results = {
        "Modelo": ["Random Forest", "AdaBoost"],
        "Acurácia Treino": [rf_train_acc, ada_train_acc],
        "Acurácia Teste": [rf_test_acc, ada_test_acc],
        "Gap (Treino - Teste)": [rf_train_acc - rf_test_acc, ada_train_acc - ada_test_acc]
    }
    
    df = pd.DataFrame(results)
    print(df.to_string(index=False))
    print("="*40)
````

- Qual modelo é mais sensível a mudanças?

O AdaBoost é o modelo mais sensível a mudanças.

Enquanto a Random Forest já entrega uma acurácia alta mesmo com poucos estimadores, o AdaBoost depende criticamente de um número elevado de estimadores e da complexidade da base para não ser um modelo "fraco". Sem esses ajustes, ele estagna em um desempenho baixo, tornando-se muito mais dependente de hiperparâmetros para funcionar no Fashion-MNIST.

````python
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

def load_data(seed=42):
    X, y = fetch_openml('Fashion-MNIST', version=1, as_frame=False, return_X_y=True, parser='auto')
    y = y.astype(int)
    return train_test_split(X, y, test_size=0.2, random_state=seed, stratify=y)

if __name__ == "__main__":
    SEED = 42
    X_train, X_test, y_train, y_test = load_data(seed=SEED)
    print("\n[8] Testando Random Forest...")
    rf_low = RandomForestClassifier(n_estimators=10, random_state=SEED, n_jobs=-1).fit(X_train, y_train)
    rf_high = RandomForestClassifier(n_estimators=150, random_state=SEED, n_jobs=-1).fit(X_train, y_train)
    print("[8] Testando AdaBoost...")
    ada_low = AdaBoostClassifier(n_estimators=10, random_state=SEED).fit(X_train, y_train)
    ada_high = AdaBoostClassifier(n_estimators=150, random_state=SEED).fit(X_train, y_train)
    
    results = [
        {"Modelo": "Random Forest", "Estimadores": 10, "Acurácia": accuracy_score(y_test, rf_low.predict(X_test))},
        {"Modelo": "Random Forest", "Estimadores": 150, "Acurácia": accuracy_score(y_test, rf_high.predict(X_test))},
        {"Modelo": "AdaBoost", "Estimadores": 10, "Acurácia": accuracy_score(y_test, ada_low.predict(X_test))},
        {"Modelo": "AdaBoost", "Estimadores": 150, "Acurácia": accuracy_score(y_test, ada_high.predict(X_test))}
    ]
    
    print("\n" + "="*45)
    print("ANÁLISE DE HIPERPARÂMETROS")
    print("="*45)
    print(pd.DataFrame(results).to_string(index=False))
````

1. A acurácia é suficiente para avaliar os modelos?
Não, a acurácia pode ser enganosa, especialmente se houvesse desbalanceamento entre as classes de roupas. Ela não diferencia tipos de erro, onde o modelo pode confundir "Camisa" com "Camiseta" ou "Tênis" com "Vestido". Métricas como Precisão, Recall e F1-Score são essenciais para entender onde o modelo falha e se ele é igualmente bom em todas as categorias.

2. Como você garante que o resultado não ocorreu por acaso?
A principal garantia vem do uso da estratificação (stratify=y) e do controle da semente de aleatoriedade (random_state). A estratificação assegura que a distribuição das classes no treino e teste seja idêntica à do dataset original. Para uma validação ainda mais robusta, o ideal seria utilizar a Validação Cruzada, que treina e testa o modelo em diferentes partições dos dados, garantindo que o desempenho seja consistente e não dependa de um sorteio de sorte.

3. Cite dois possíveis problemas metodológicos neste experimento.
O primeiro problema é o Underfitting do AdaBoost devido ao uso do estimador padrão (árvores de profundidade 1), o que impossibilita uma comparação justa de "poder máximo" contra a Random Forest. O segundo é a falta de um Conjunto de Validação distinto do conjunto de Teste; ao ajustar hiperparâmetros (como o n_estimators na Questão 8) baseando-se apenas no resultado do Teste, corremos o risco de "vazar" informação e superestimar a capacidade de generalização do modelo final.

4. O pipeline implementado é confiável? Justifique.
Sim, o pipeline é confiável do ponto de vista de reprodutibilidade e organização. Ele isola as etapas de carga, treino e avaliação, evitando que dados de teste "vazem" para o treinamento e garantindo que qualquer pessoa com o mesmo código e semente obtenha resultados idênticos. No entanto, para ser considerado um pipeline de produção robusto, ele precisaria incluir etapas de tratamento de erros (ex: falha no download do dataset) e uma busca exaustiva de parâmetros (GridSearch).
